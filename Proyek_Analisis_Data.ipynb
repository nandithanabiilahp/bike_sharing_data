## Visualization & Explanatory Analysis

Di hari apa rata-rata penggunaan sepeda paling tinggi, dan di hari apa paling rendah?

```python
# Simulasikan DataFrame day_df
# Gantilah dengan data yang sesuai dengan struktur DataFrame Anda
day_df = pd.DataFrame({
    'day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
    'bike_count': [100, 120, 90, 150, 80]
})

# Urutkan DataFrame berdasarkan rata-rata penggunaan sepeda
sorted_day_df = day_df.sort_values(by='bike_count', ascending=False)

# Visualisasi bar chart untuk menunjukkan rata-rata penggunaan sepeda
plt.figure(figsize=(10, 6))
plt.bar(sorted_day_df['day'], sorted_day_df['bike_count'], color='skyblue')
plt.title('Average Bike Usage by Day')
plt.xlabel('Day')
plt.ylabel('Average Bike Count')
plt.show()
```

### Pertanyaan 2: Apakah ada hari yang memiliki median (nilai tengah) penggunaan sepeda yang berbeda signifikan dari rata-rata?

```python
# Simulasikan DataFrame day_df
# Gantilah dengan data yang sesuai dengan struktur DataFrame Anda
day_df = pd.DataFrame({
    'day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
    'bike_count': [100, 120, 90, 150, 80]
})

# Visualisasi boxplot untuk menunjukkan distribusi penggunaan sepeda per hari
plt.figure(figsize=(10, 6))
plt.boxplot(day_df['bike_count'], vert=False, labels=['Bike Usage'])
plt.title('Distribution of Bike Usage by Day')
plt.xlabel('Bike Count')
plt.show()
```

## Conclusion

- Conclution pertanyaan 1 : The code generates a bar chart to visualize the average bike usage for each day in the day_df DataFrame. It allows us to quickly identify the day with the highest and lowest average bike usage.
- Conclution pertanyaan 2 : The code uses a boxplot to compare the distribution of bike usage across different days in the day_df DataFrame. It helps assess whether there are significant differences between the median and mean bike usage for each day, indicating potential distribution skewness. The visualization provides insights into the symmetry of the data distribution.
